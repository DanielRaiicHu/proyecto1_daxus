## Generando archivos PDF del presupuesto
* Crear el archivo PDF
* Definir un tipo de letra
* Llenamos el documento
* Guardar el entregable


```python
# Librería para instalar PDF
!pip install fpdf
```

```python
from fpdf import FPDF
```

```python
pdf = FPDF()
pdf.add_page() # Agregar página
pdf.set_font("Arial") # Seleccionar tipo de fuente
pdf.image("Template.png", 0, 0) # Imagen de fondo
# Atento con las coordenadas en el archivo del PDF (x,y,texto)
pdf.text(115,145, proyecto)
pdf.text(115,160, horas_estimadas)
pdf.text(115,175, valor_hora)
pdf.text(115,190, plazo_estimado)
pdf.text(115,205, str(valor_total))
pdf.output("Presupuesto.pdf") # Crear archivo
```
